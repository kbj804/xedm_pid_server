```python
import os
file_list = os.listdir('./csv_files')
len(file_list)
```

```python
def show_data(data_path):
    train = pd.read_csv(f'./csv_files/{data_path}')
    if train[train['x'] > 20].values.tolist():
        return 1
    else:
        return 0

trash_list = []

for file in file_list:
    flag = show_data(file)
    if flag == 1:
        trash_list.append(file)
    else:
        pass
```

```python
len(trash_list)
```

```python
test = pd.read_csv('./submission.csv')
test
```

```python
def change(f_name):
    if f_name in trash_list:
        return 1
    else:
        return 0

test['anomaly_score'] = test['rand_filename'].apply(lambda x : change(x))

    
```
